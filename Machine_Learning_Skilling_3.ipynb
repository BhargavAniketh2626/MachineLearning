#Machine Learning Skilling 3

##Importing of the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##Importing of the dataset and perform slicing operations for X and Y seperations

```python
df = pd.read_csv('php0iVrYT.csv')
X = df.iloc[:, 2:3].values
y = df.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

#Training the Linear Regression model on whole dataset

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
```

##Converting of the y dataset into columns

```python
y = y.reshape(len(y), 1)
```

```python
y
```

##Splitting the training and testing dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/5, random_state = 0)
```

##Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

##Predict the value

```python
y = lr.predict(X_test)
```

##Visualizing on the training resultset

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, lr.predict(X_train), color='blue')
plt.title("Salaries vs Experience (Testing result)")
plt.xlabel('Salaries')
plt.ylabel('Experience')
plt.show()
```

```python
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, lr.predict(X_test), color='blue')
plt.title("Salaries vs Experience (Testing result)")
plt.xlabel('Salaries')
plt.ylabel('Experience')
plt.show()
```
